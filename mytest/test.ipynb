```python
import pandas as pd
import numpy as np

# 创建示例 DataFrame
df = pd.DataFrame({
    'A': np.random.randn(100),
    'B': np.random.randn(100) * 2,
    'C': np.random.randint(0, 10, 100),  # 整数列，不会被选中
    'D': np.random.randn(100) + 5,
    'E': ['x', 'y'] * 50                 # 字符串列
})

# 筛选 float 列并绘制直方图
float_cols = df.select_dtypes(include='float')
float_cols.hist(figsize=(10, 6), bins=20)

# 显示图形（如果在 Jupyter 中自动显示，否则需 plt.show()）
import matplotlib.pyplot as plt
plt.show()
```
